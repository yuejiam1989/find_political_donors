```python
import numpy as np
import pandas as pd
import datetime, calendar
from compiler.ast import flatten
```

```python
## read data, make it into data frame

data = pd.read_csv(r"C:\Users\Sissi\Downloads\find-political-donors-master\input\itcont_2016_20161214_92060702.txt", 
                   header = None, delimiter = "|", dtype = object)
## "C:\Users\Sissi\Downloads\find-political-donors-master\insight_testsuite\tests\test_1\input\itcont.txt"
# df = pd.DataFrame(data)
# display(df)
data = data.head(1000)
# display(data)
```

```python
## extract useful information (5 columns), rename the columns

udata = data[[0, 10, 13, 14, 15]]
headline = ["CMTE_ID", "ZIP_CODE", "TRANSACTION_DT", "TRANSACTION_AMT", "OTHER_ID"]
udata.columns = headline
# udata = df["CMTE_ID", "ZIP_CODE", "TRANSACTION_DT", "TRANSACTION_AMT", "OTHER_ID"]
# display(udata)
```

```python
## clean data. delete null tran_amt, null cmte_ID and non-empty other_ID and

pd.to_numeric(udata["TRANSACTION_AMT"], errors='coerce')
udata = udata[pd.notnull(udata["TRANSACTION_AMT"])]
udata.dropna(subset=["CMTE_ID"], inplace = True)
udata = udata[pd.isnull(udata["OTHER_ID"])]
# display(udata)
```

```python
### task 1: medianvals_by_zip.txt

## change ZIP to 5 digital, otherwise, (empty or invalid), drop it
zd = udata.copy()
zd["ZIP_CODE"]= zd.ZIP_CODE.str[:5]
# display(zd)
# display(udata)

zd = zd[zd.ZIP_CODE.str.len() == 5]
# display(zd)
```

```python
## change TRANSACTION_AMT data type to number.

# udata.TRANSACTION_AMT = udata.TRANSACTION_AMT.apply(int)
zd.TRANSACTION_AMT = zd.TRANSACTION_AMT.apply(int)
```

```python
## set original lists for loop

TN_TRAN = [1] * len(zd)
TA_TRAN = [0] * len(zd)
LTA_TRAN = [[] for i in range(len(zd))]
RMEDIAN_ZIP = [0] * len(zd)
```

```python
## loop, caculate No., total amt, median.

for i, idx in enumerate(zd.index):
    TA_TRAN[i] = zd.TRANSACTION_AMT[idx]
    LTA_TRAN[i] = [zd.TRANSACTION_AMT[idx]]
    RMEDIAN_ZIP[i] = round(zd.TRANSACTION_AMT[idx], 0)
#   print(TN_TRAN[i], TA_TRAN[i], LTA_TRAN[i], RMEDIAN_ZIP[i])
    for l, ld in reversed(list(enumerate(zd.index))):
            if (ld < idx) & (zd.ZIP_CODE[idx] == zd.ZIP_CODE[ld]):
                TN_TRAN[i] = TN_TRAN[l] + 1
                TA_TRAN[i] = zd.TRANSACTION_AMT[idx] + TA_TRAN[l]
                LTA_TRAN[i].append(flatten([LTA_TRAN[l]]))
                RMEDIAN_ZIP[i] = round(np.median(flatten(LTA_TRAN[i])) , 0)
#               print(TN_TRAN[i], TA_TRAN[i], LTA_TRAN[i], RMEDIAN_ZIP[i])
                break

RMEDIAN_ZIP = [int(i) for i in RMEDIAN_ZIP]    
z = pd.DataFrame(
    {
     "1": zd.CMTE_ID,
     "2": zd.ZIP_CODE,
     "3": RMEDIAN_ZIP,
     "4": TN_TRAN,
     "5": TA_TRAN
    })
headline = ["CMTE_ID", "ZIP_CODE", "RMEDIAN_ZIP", "TOTNUMBER_TRAN", "TOTAMT_TRAN"]
z.columns = headline

display(z)
```

```python
## save as text file

np.savetxt(r"C:\Users\Sissi\Downloads\find-political-donors-master\input\medianvals_by_zip.txt", 
           z.values, delimiter = "|", fmt="%s")
# r"C:\Users\Sissi\Downloads\find-political-donors-master\insight_testsuite\tests\test_1\input\medianvals_by_zip.txt"
```

```python
### task 2: medianvals_by_date.txt
```

```python
## change TRANSACTION_DT to datetime form and verify the date. If date is invalid, drop it

td = udata.copy()
td["TRANSACTION_DT"]= pd.to_datetime(udata.TRANSACTION_DT, format='%m%d%Y',errors='coerce')
# display(td)
# display(udata)

td = td[pd.notnull(td["TRANSACTION_DT"])]
td["TRANSACTION_DT"] = td["TRANSACTION_DT"].apply(lambda x: x.strftime("%m%d%Y"))
display(td)
```

```python
## change TRANSACTION_AMT data type to number.

td.TRANSACTION_AMT = td.TRANSACTION_AMT.apply(int)
```

```python
## Calculate Median, sum and count of Group over Columns

res = td.groupby(["CMTE_ID", "TRANSACTION_DT"])["TRANSACTION_AMT"].agg(['median', 'count', 'sum'])
```

```python
## change the type of each columns, made into dataframe

gb = td.groupby(["CMTE_ID", "TRANSACTION_DT"])
groupn = gb.apply(lambda x: x.name)
id = groupn.index.get_level_values(0)
date = groupn.index.get_level_values(1)
ml = res.values[:,0].tolist()
mmedian = [int(round(elem, 0)) for elem in ml]
cd = res.values[:,1].tolist()
mcd = [int(i) for i in cd]
sd = res.values[:,2].tolist()
msd = [int(i) for i in sd]

d = pd.DataFrame(
    {
     "1": id,
     "2": date,
     "3": mmedian, 
     "4": mcd,
     "5": msd   
    })
headline = ["CMTE_ID", "TRANSACTION_DT", "MEDIAN_DT", "TOTNUMBER_DT", "TOTAMT_DT"]
d.columns = headline

display(d)
```

```python
## save as text file

np.savetxt(r"C:\Users\Sissi\Downloads\find-political-donors-master\input\medianvals_by_date.txt", 
           d.values, delimiter = "|", fmt="%s")
# r"C:\Users\Sissi\Downloads\find-political-donors-master\insight_testsuite\tests\test_1\input\medianvals_by_date.txt"
```
